```python
import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,None,None],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,95,96,66]
}
df=pd.DataFrame(data)
df=df.fillna(0)
print(df)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,None,None],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,95,96,66]
}
df=pd.DataFrame(data)
df["Marks"]=df["Marks"].fillna(df["Marks"].mean())
print("\nAfter filling missing Marks :")
print(df)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,None,None],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,95,96,66]
}
df=pd.DataFrame(data)
df["Age"]=df["Age"].fillna(df["Age"].mean())
print("\nAfter filling missing age :")
print(df)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,None,None],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,95,96,66]
}
df=pd.DataFrame(data)
df=df.dropna()
print("\nAfter removing rows with missing values:")
print(df)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,None,None],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
df=df.duplicated()
print("\nDuplicate records:")
print(df)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,None,None],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
df=df.drop_duplicates()
print("\nAfter removing duplicate records:")
print(df)


import pandas as pd

data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,None,None],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
result=df[(df["Marks"]>df["Marks"].mean())]
print("\nStudents whose marks are greater than average marks:")
print(result)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,None,None],

    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
att_result=df[(df["Attendance"]>90)]
print("\nStudents whose attendance is greater than 90%:")
print(att_result)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,None,None],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
marks_result=df[(df["Marks"]>=70) & (df["Marks"]<=90)]
print("\nStudents whose marks are between 70 and 90:")

print(marks_result)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
print(df)
result=df[(df["Gender"]=="Female") & (df["City"]=="Pune")]
print(result)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
result=df[(df["Gender"]=="Male") & (df["Marks"]>80)]
print("\nMale students whose marks are greater than 80:")
print(result)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
result=df["Marks"].max()
print("\nMaximum marks of students:")
print(result)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],

    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
result=df["Marks"].min()
print("\nMinimum marks of students:")
print(result)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
result=df["Marks"].mean()
print("\nAverage marks of all students:")
print(result)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
result=df["Marks"].sum()
print("\nTotal marks of all students:")
print(result)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66],
    "Course":["AIML","Computer","Civil","Computer","Mechanical"]
}
df=pd.DataFrame(data)
result=df["Course"].unique()
print("\n Number of unique course:")
print(result)


import pandas as pd

data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66],
    "Course":["AIML","Computer","Civil","Computer","Mechanical"]
}
df=pd.DataFrame(data)
result=df["City"].value_counts()
print("\n Number of students in each city:")
print(result)


import pandas as pd
data={
   
 "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66],
    "Course":["AIML","Computer","Civil","Computer","Mechanical"]
}
df=pd.DataFrame(data)
result=df["Course"].value_counts()
print("\nNumber of students in each course:")
print(result)


import pandas as pd
data={
  
  "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66],
    "Course":["AIML","Computer","Civil","Computer","Mechanical"]
}
df=pd.DataFrame(data)
average_marks=df.groupby("City")["Marks"].mean()
print(average_marks)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66],
    "Course":["AIML","Computer","Civil","Computer","Mechanical"]
}
df=pd.DataFrame(data)
average_marks=df.groupby("Course")["Marks"].agg(["sum","mean","min","max","count"])
print(average_marks)


import pandas as pd
data={
    "Age":[18,None,18,19,None],
    "Marks":[88,78,88,88,89],
    "Attendance":[90,80,90,96,66]
}
df=pd.DataFrame(data)
print(df)
correlation=df.corr()
print(correlation)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    
"City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66],
    "Course":["AIML","Computer","Civil","Computer","Mechanical"]
}
df=pd.DataFrame(data)
df["Gender"]=df["Gender"].replace({"Male":1,"Female":0})
print(df)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66],
    "Course":["AIML","Computer","Civil","Computer","Mechanical"],
    "Purchased":[1,0,1,0,0]
}
df=pd.DataFrame(data)
encoded_df=pd.get_dummies(df,columns=["City","Course"])
print(encoded_df)


import pandas as pd
data={
    "Name":["Priya","Om","Priya","Rohan","Rahul"],
    "Age":[18,None,18,19,None],
    "Gender":["Female","Male","Female","Male","Male"],
    "Marks":[88,78,88,88,89],
    "City":["Pune","Mumbai","Pune","Delhi","Pune"],
    "Attendance":[90,80,90,96,66],
    "Course":["AIML","Computer","Civil","Computer","Mechanical"],
    "Purchased":[1,0,1,0,0]
}
df=pd.DataFrame(data)
X=df[["Age","Attendance","Marks"]]
y=df["Purchased"]
print("Featrues (X):")

print(X)
print("\nTarget (y):")
print(y)


import pandas as pd

# Read the CSV file
df = pd.read_csv("students.csv")

# Display the dataset
print("Dataset:")
print(df)

# Display first 5 rows
print("\nFirst 5 Rows:")
print(df.head())

# Display last 5 rows
print("\nLast 5 Rows:")
print(df.tail())

# Display dataset information
print("\nDataset Information:")
print(df.info())

# Display statistical summary
print("\nStatistical Summary:")
print(df.describe())

# Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Handle missing values
df['Marks'].fillna(df['Marks'].mean(), inplace=True)

# Find duplicate records
print("\nDuplicate Records:")
print(df.duplicated())

# Remove duplicate records
df = df.drop_duplicates()


# Filter records
print("\nStudents with Marks > 75:")
print(df[df['Marks'] > 75])

print("\nStudents from Pune:")
print(df[df['City'] == 'Pune'])

# Sort the dataset
print("\nSorted by Marks:")
print(df.sort_values(by='Marks', ascending=False))

# Display unique values
print("\nUnique Cities:")
print(df['City'].unique())

# Count values
print("\nCity Count:")
print(df['City'].value_counts())

# GroupBy analysis
print("\nAverage Marks by City:")
print(df.groupby('City')['Marks'].mean())

# Aggregation functions
print("\nAggregation:")
print(df['Marks'].agg(['sum', 'mean', 'min', 'max', 'count']))

# Correlation
print("\nCorrelation:")
print(df.corr(numeric_only=True))


# Manual Encoding
#df['Gender'] = df['Gender'].replace({'Male': 1, 'Female': 0})

# One-Hot Encoding
df = pd.get_dummies(df, columns=['City'])

# Create Features (X) and Target (y)
X = df[['Age','Marks']]
y = df['Marks']

print("\nFeatures (X):")
print(X)

print("\nTarget (y):")
print(y)

# Save cleaned dataset
df.to_csv("cleaned_students.csv", index=False)

print("\nCleaned dataset saved successfully as cleaned_students.csv")
```
